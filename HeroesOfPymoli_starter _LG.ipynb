```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Save file path to variable "pymoli_path"
pymoli_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame "pur_data"
pur_data = pd.read_csv(pymoli_path)
pur_data.head()
```

## Player Count

* Display the total number of players


```python
# Slim pur_data to only contain columns of "SN", "Age", and "Gender" and save it as "player_info"
player_info = pur_data[["SN", "Age", "Gender"]]

# Drop duplicated rows in "player_info" 
player_info = player_info.drop_duplicates()

# Create a DataFrame called "total_player" to store the counts of "SN" in "player_info"
total_player = pd.DataFrame(
    {"Total Players": [player_info["SN"].count()] })

total_player
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Create a DataFrame to analyze purchase behaviors  
pur_glance = pd.DataFrame(
    {"Number of Unique Items": [len(pur_data["Item ID"].unique())],
     "Average Price": [pur_data["Price"].mean()],
     "Number of Purchases": [pur_data["Price"].count()],
     "Total Revenue": [pur_data["Price"].sum()]
    })

# Format "Average Price" and "Total Revenue" columns, respectively
pur_glance["Average Price"] = pur_glance["Average Price"].map("${:.2f}".format)
pur_glance["Total Revenue"] = pur_glance["Total Revenue"].map("${:,}".format)

pur_glance
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Save counts of each gender from "player_info" in a Series called "gender_count"
gender_count = player_info["Gender"].value_counts()
```

```python
# Calculate and save player's count of  gender  percentage in a Series called "gender_pct"

gender_pct = player_info["Gender"].value_counts() * 100/player_info["SN"].count()
```

```python
# Create a dataframe
gender_info = pd.DataFrame({"Total Count": gender_count,
    "Percentage of Players": gender_pct})
```

```python
# Format value in "Percentage of Players" column
gender_info["Percentage of Players"] = gender_info["Percentage of Players"].map("{:.2f}".format)

gender_info
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Use GroupBy in order to separate the original "pur_data" into fields according to "Gender"
grouped_gender_pur = pur_data.groupby(["Gender"], as_index = True)

# Create a new DataFrame called "pur_gender"
# Calculate and save "Purchase Count", "Average Purchase Price", "Total Purchase Value", and "Avg Total Purchase per Person" in "pur_gender"
pur_gender = pd.DataFrame(
    {"Purchase Count": grouped_gender_pur["Purchase ID"].count(),
     "Average Purchase Price": grouped_gender_pur["Price"].mean(),
     "Total Purchase Value": round(grouped_gender_pur["Price"].sum(), 2),
     "Avg Total Purchase per Person": grouped_gender_pur["Price"].sum() / gender_info["Total Count"]
    })

# Format "Average Purchase Price", "Total Purchase Value" and "Avg Total Purchase per Person" columns, respectively
pur_gender["Average Purchase Price"] = pur_gender["Average Purchase Price"].map("${:.2f}".format)
pur_gender["Total Purchase Value"] = pur_gender["Total Purchase Value"].map("${:,}".format)
pur_gender["Avg Total Purchase per Person"] = pur_gender["Avg Total Purchase per Person"].map("${:.2f}".format)


pur_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Set up the bins for "player_info"
# Assume people with an age of 100+ won't be interested in gaming ... 
bins=[0,9,14,19,24,29,34,39,50]

# Create the names for binned categories
age_range = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice the data by binning "Age" column of "player_info" and place the data series into a new column called "Age Group"  
player_info["Age Group"] = pd.cut(player_info["Age"], bins,labels=age_range)
pur_data["Age Group"] = pd.cut(pur_data["Age"], bins,labels=age_range)
player_info.head()
```

```python
# Sort "player_info" based on "Age Group" column
player_info = player_info.sort_values("Age Group")

# Calculate the numbers by age group
age_count = player_info["Age Group"].value_counts()

# Calculate percentages of player's age group 
age_pct = player_info["Age Group"].value_counts() * 100/ len(player_info)
```

```python

# Create a new DataFrame called "age_info"
age_info = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_pct})

# Format "Percentage of Players"
age_info["Percentage of Players"] = age_info["Percentage of Players"].map("{:.2f}".format)

# Sort rows by index
age_info.sort_index(axis=0,  ascending=True)
```

## Purchasing Analysis (Age)

* Bin the pur_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Calculate the purchase value by age group
Total_value_per_group = pur_data.groupby('Age Group').Price.sum().values

# Calculate the average purchase value by age group and  save the values as a list
average_purch=pur_data.groupby('Age Group').Price.mean().values

# Calculate the purchase times by age group and save the values as a list
purch_count = pur_data.groupby('Age Group').Price.count().values                    

# Calculate the Average total purchase per person and save the values as a list
# Average total purchase per person = total purchase value in each age group/ total number of player in each age group
avg_purch_per_person=pur_data.groupby('Age Group').Price.sum().values/age_count.sort_index(axis=0,  ascending=True).values
```

```python
# Convert all data to a dataframe
raw_data_info = {'Purchase Count':purch_count,'Average Purchase Price':average_purch,
                  'Total Purchase Value':Total_value_per_group,
                 'Avg Total Purchase per Person':avg_purch_per_person}

df_columns=['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']
df_index=list(age_count.sort_index(axis=0,  ascending=True).keys())
purchase_age_df = pd.DataFrame(raw_data_info,columns=df_columns,index=df_index)
```

```python
# Format the value in 'Average Purchase Price','Total Purchase Value',
# 'Avg Total Purchase per Person' columns
columns=['Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']
for c in columns:
    purchase_age_df[c] = purchase_age_df[c].map("${:.2f}".format)    
   
```

```python
# Organize the rows, put the first row(index="<10") to the last one
purchase_age_df=purchase_age_df[1:].append(purchase_age_df[:1])
purchase_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculate the purchase times by SN group
Purchase_count = pd.DataFrame(pur_data.groupby('SN').count().Price)
# Rename the column
Purchase_count = Purchase_count.rename(columns={'Price':'Purchase Count'})
```

```python
# Calculate the total purchase value by SN group
total_value = pd.DataFrame(pur_data.groupby('SN').sum().Price)
# Rename the column
total_value = total_value.rename(columns={'Price':'Total Purchase Value'})
```

```python
# Combine two dataframe Purchase_count and total_value

top_spender_df = Purchase_count.join(total_value,how='left')
```

```python
# Calculate the average purchase price
top_spender_df['Average Purchase Price']= top_spender_df['Total Purchase Value']/top_spender_df['Purchase Count']
top_spender_df.head()
```

```python
# Sort the total purchase value column in descending order
# Slice the dataframe only keep the first five rows
top_spender_df = top_spender_df.sort_values(by=['Total Purchase Value'],ascending=False)[:5]
```

```python
# Reorgnize the columns and format the dataframe 
top_spender_df = top_spender_df[['Purchase Count','Average Purchase Price','Total Purchase Value']]
top_spender_df['Total Purchase Value']=top_spender_df['Total Purchase Value'].map('${:.2f}'.format)
top_spender_df['Average Purchase Price']=top_spender_df['Average Purchase Price'].map('${:.2f}'.format)
top_spender_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns

* Create a summary data frame to hold the results


* Sort the purchase count column in descending order

```python
# Retrieve the Item ID, Item Name, and Item Price columns
Popular_Items_df=pur_data[['Item ID','Item Name','Price']]
Popular_Items_df.head()
```

```python
# Calculate the total purchase value by 'Item ID','Item Name' columns
Popular_Items_df= Popular_Items_df.groupby(['Item ID','Item Name'])['Price'].value_counts()
# Convert to dataframe
Popular_Items_df= pd.DataFrame(Popular_Items_df)
# Remane column and convert the Price index to column
Popular_Items_df = Popular_Items_df.rename(columns={'Price':'Purchase Count'})
Popular_Items_df = Popular_Items_df.reset_index(level=['Price'])
```

```python
# Remane column
Popular_Items_df = Popular_Items_df.rename(columns={'Price':'Item Price'})
Popular_Items_df.head()
```

```python
# Check the puchase count value distribution
Popular_Items_df['Purchase Count'].describe()
```

```python
# Calculate the total purchase value
Popular_Items_df['Total Purchase Value']= (Popular_Items_df['Item Price'])* (Popular_Items_df['Purchase Count'])
```

```python
# Format the value
Popular_Items_df['Item Price']=Popular_Items_df['Item Price'].map("${:.2f}".format)    
Popular_Items_df['Total Purchase Value']=Popular_Items_df['Total Purchase Value'].map("${:.2f}".format)
```

```python
# Reorganize the columns name
Popular_Items_df= Popular_Items_df[['Purchase Count','Item Price','Total Purchase Value']]
# Sort the purchase count column in descending order
Popular_Items_df= Popular_Items_df. sort_values(by='Purchase Count', ascending=False)
Popular_Items_df.head()
```
